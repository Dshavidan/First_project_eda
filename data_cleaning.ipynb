```python
# Datenabfrage bei DB Beaver um sich eine Übersicht zu erhalten

'''

-- ##EDA Project

SELECT 
    s.house_id,
    s.price,
    s.date,
    d.bedrooms,
    d.bathrooms,
    d.sqft_living,
    d.waterfront
FROM eda.king_county_house_sales AS s
JOIN eda.king_county_house_details AS d
ON s.house_id = d.id
LIMIT 10000;

###Tabellenüberbblick: Spalten und Struktur

SELECT * FROM eda.king_county_house_sales LIMIT 100;
SELECT * FROM eda.king_county_house_details LIMIT 100;

###Tabellen verbinden mit der JOIN Funktion 

SELECT *
FROM eda.king_county_house_sales AS s
JOIN eda.king_county_house_details AS d
ON s.house_id = d.id;

###Durchschnittlicher Verkauspreis pro ZIP Code (Hypothese 1)

SELECT d.zipcode, ROUND(AVG(s.price)) AS avg_price
FROM eda.king_county_house_sales s
JOIN eda.king_county_house_details d ON s.house_id = d.id
GROUP BY d.zipcode
ORDER BY avg_price ASC;  ### Durschnittliche Verkaufspreis beträgt den 98. ZIP Code Reihen zwischen 200k bis 1 Mio 

### Durchschnittlicher Baujahr pro ZIP Code (Hypothese 2)


SELECT d.zipcode, ROUND(AVG(d.yr_built)) AS avg_year_built
FROM eda.king_county_house_sales s
JOIN eda.king_county_house_details d ON s.house_id = d.id
GROUP BY d.zipcode
ORDER BY avg_year_built ASC; ### Durchschnittliche Baujahr beträgt zwischen 1940  bis 1998

### Häsuer in bestimmmten ArmenZonen finden 

SELECT s.price, d.zipcode, d.yr_built, d.bedrooms, d.bathrooms
FROM eda.king_county_house_sales s
JOIN eda.king_county_house_details d ON s.house_id = d.id
WHERE d.zipcode IN (98188, 98002, 98001, 98168, 98003)
ORDER BY s.price ASC;
'''

```

```python
# Hypothese 1

import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker

avg_price_zip = df_kingcounty.groupby("zipcode")["price"].mean().sort_values().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=avg_price_zip.head(20), x="zipcode", y="price")
plt.title("Average Price in 20 Poorest Zipcodes")
plt.xlabel("Zipcode")
plt.ylabel("Average Price (USD)")
plt.xticks(rotation=45)
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'${x:,.0f}'))
plt.tight_layout()
plt.show()

# Dieser Plot zeigt die 20 günstigsten Durchschnittlichen Preise im Bereich der 20 günstigsten Areas.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Durchschnittspreis pro PLZ berechnen
avg_price_zip = df_kingcounty.groupby("zipcode")["price"].mean().sort_values()

# 20 günstigste und 20 teuerste Zipcodes
poorest_zipcodes = avg_price_zip.head(20).index
richest_zipcodes = avg_price_zip.tail(20).index

# Neue Spalte: Kategorie „arm“ oder „reich“
df_kingcounty["neighborhood_type"] = df_kingcounty["zipcode"].apply(
    lambda x: "Poorest 20" if x in poorest_zipcodes else ("Richest 20" if x in richest_zipcodes else "Other")
)

# Nur "Poorest 20" und "Richest 20" behalten
subset = df_kingcounty[df_kingcounty["neighborhood_type"].isin(["Poorest 20", "Richest 20"])]

# Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=subset, x="neighborhood_type", y="price")
plt.title("House Price Distribution: Poorest vs. Richest Neighborhoods")
plt.xlabel("Neighborhood Type")
plt.ylabel("House Price (USD)")
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'${x:,.0f}'))
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Durchschnittspreis pro Zipcode berechnen
avg_price_zip = df_kingcounty.groupby("zipcode")["price"].mean().rename("avg_zip_price")

# Durchschnittspreis in das ursprüngliche DataFrame mergen
df_scatter = df_kingcounty.merge(avg_price_zip, on="zipcode")

# Streudiagramm erstellen
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_scatter, x="avg_zip_price", y="price", alpha=0.3)

plt.title("House Price vs. Average Price of Zipcode")
plt.xlabel("Average Price in Zipcode (Proxy for Neighborhood Wealth)")
plt.ylabel("House Price")
plt.gca().xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'${x:,.0f}'))
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'${x:,.0f}'))

plt.tight_layout()
plt.show()

# Steigt der Hauspreis mit dem durchschnittlichen Zipcode-Preis, stützt das die Hypothese:
# → „Häuser in ärmeren Vierteln (mit niedrigerem Zipcode-Durchschnitt) sind günstiger.“

#Viele Ausreißer nach oben/unten? → Dann lohnt sich ein Blick auf weitere Einflussfaktoren wie Größe oder Zustand.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker

# Durchschnittspreis pro Zipcode berechnen
avg_price_zip = df_kingcounty.groupby("zipcode")["price"].mean().rename("avg_zip_price")

# DataFrame mit Zipcode-Durchschnittspreis erweitern
df_scatter = df_kingcounty.merge(avg_price_zip, on="zipcode")

# Plot mit Trendlinie
plt.figure(figsize=(10, 6))
sns.regplot(data=df_scatter, x="avg_zip_price", y="price", scatter_kws={"alpha": 0.3}, line_kws={"color": "red"})

plt.title("House Price vs. Average Price of Zipcode (with Trendline)")
plt.xlabel("Average Price in Zipcode (Neighborhood Wealth)")
plt.ylabel("House Price")

# Formatierung der Achsen (optional, für bessere Lesbarkeit)
plt.gca().xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'${x:,.0f}'))
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'${x:,.0f}'))

plt.tight_layout()
plt.show()

#🔍 Interpretation der Trendlinie:

#Wenn die Trendlinie ansteigt, deutet das auf einen positiven Zusammenhang hin → Häuser sind in wohlhabenderen Gegenden teurer.
#Damit wird die Hypothese „Häuser in ärmeren Vierteln sind günstiger“ unterstützt.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.patches import Patch
from matplotlib.lines import Line2D

# Durchschnittspreis pro Zipcode berechnen
avg_price_per_zip = df_kingcounty.groupby("zipcode")["price"].mean()
overall_avg_price = avg_price_per_zip.mean()

# Klassifizierung in "poor" und "rich"
zip_class = avg_price_per_zip.apply(lambda x: "poor" if x < overall_avg_price else "rich")
df_hist = df_kingcounty.copy()
df_hist["zip_class"] = df_hist["zipcode"].map(zip_class)

# Preis in Millionen umrechnen
df_hist["price_million"] = df_hist["price"] / 1_000_000

# Stil setzen
sns.set(style="whitegrid")

# Plot vorbereiten
plt.figure(figsize=(12, 6))

# Histogramm + KDE
sns.histplot(
    data=df_hist[df_hist["zip_class"] == "poor"],
    x="price_million",
    bins=50,
    kde=True,
    color="blue",
    label="Poor - Histogramm",
    alpha=0.5
)

sns.histplot(
    data=df_hist[df_hist["zip_class"] == "rich"],
    x="price_million",
    bins=50,
    kde=True,
    color="orange",
    label="Rich - Histogramm",
    alpha=0.5
)

# Legende manuell erstellen mit Kästchen und Linie
legend_elements = [
    Patch(facecolor="blue", alpha=0.5, label="Poor - Histogramm"),
    Line2D([0], [0], color="blue", lw=2, label="Poor - KDE-Kurve"),
    Patch(facecolor="orange", alpha=0.5, label="Rich - Histogramm"),
    Line2D([0], [0], color="orange", lw=2, label="Rich - KDE-Kurve"),
]

plt.legend(handles=legend_elements, title="Vierteltyp")

# Titel und Achsen
plt.title("Hauspreisverteilung in ärmeren vs. reicheren Gegenden", fontsize=14)
plt.xlabel("Hauspreis (in Millionen USD)", fontsize=12)
plt.ylabel("Anzahl der Häuser", fontsize=12)
plt.xlim(0, 2)
plt.tight_layout()
plt.show()

plt.show()
```
